```python
i = 1
# create a loop running if "data" is not None
while True:
    data = server.receive_data()
    if data is not None:
        clear_output(wait=True)
        print("Iteration: ", i)
        print("Received data: ", data)
        i += 1
    else:
        clear_output(wait=True)
        print("Data is None: ", [data, i])
        i += 1
    time.sleep(0.1)
```
